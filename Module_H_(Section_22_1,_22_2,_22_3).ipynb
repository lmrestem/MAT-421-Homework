```python
# Euler Method Forward example
import numpy as np

def forward_euler(f, U0, T, n):
  t = np.zeros(n + 1)
  u = np.zeros(n + 1)
  u[0] = U0
  dt = T / float(n)
  for k in range(n):
    t[k + 1] = t[k] + dt
    u[k + 1] = u[k] + dt*f(u[k], t[k])
  return u, t

from matplotlib import pyplot as plt
def f(u, t):
    return u
  
for n in [5, 10, 20]:
    u, t = forward_euler(f, U0=1, T=4, n=n)
    plt.plot(t, u, linestyle="dashed", label=f"n={n}")

t_fine = np.linspace(0, 4, 1001)
plt.plot(t_fine, np.exp(t_fine), label="Exact")
plt.legend()
plt.show()
print('')

# Euler method example
def fun(x, y):
  return (x + y + x *y)

def euler(x0, y, h, x):
  temp = -0
  while x0 < x:
    temp = y
    y = y + h * fun(x0, y)
    x0 = x0 + h
  print("Approximate solution using Euler Method at x = ", x, " is ", "%.6f"% y)

x0 = 0
y0 = 1
h = 0.2
x = 0.5

euler(x0, y0, h, x)
```
