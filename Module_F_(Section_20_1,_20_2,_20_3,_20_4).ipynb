```python
# taylor series approximation example

import sympy as sy
import numpy as np
from sympy.functions import sin,cos
import matplotlib.pyplot as plt

plt.style.use("ggplot")

x = sy.Symbol('x')
f = sin(x)

def factorial(n):
  if n <= 0:
    return 1 
  else: 
    return n * factorial(n-1)

def taylor(function, x0, n):
  i = 0
  p = 0
  while i <= n:
    p = p + (function.diff(x, i).subs(x, x0)) / (factorial(i)) * (x - x0) ** i
    i += 1
  return p 

def plot():
  x_lims = [-6, 6]
  x1 = np.linspace(x_lims[0], x_lims[1], 800)
  y1 = []
  for j in range(1, 10, 2):
    function = taylor(f, 0, j)
    print('Taylor expansion at n =' +str(j), function)
    for k in x1:
      y1.append(function.subs(x, k))
    plt.plot(x1, y1, label = 'order ' +str(j))
    y1 = []
    plt.plot(x1, np.sin(x1), label = 'sin of x')
    plt.xlim(x_lims)
    plt.ylim([-6, 6])
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid(True)
    plt.title('Taylor Series Approximation')
    plt.show()

plot()
print('')


# approximating second derivative example

def f(x):
  return 2 * x * x + 10

def d2(fn, x0, h):
  return (fn(x0 + h) - 2 * fn(x0) + fn(x0 - h)) / (h * h)

print('Approximation of second derivative')
print(d2(f, 12, 0.1))
```
